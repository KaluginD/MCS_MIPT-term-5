T(N)

```python
thread4_exp1 = '0.50 2412.1 0.0153s 0 100 50 500 0.50 4'
thread4_exp2 = '0.50 2537.1 0.0256s 0 100 50 1000 0.50 4'
thread4_exp3 = '0.51 2471.9 0.0386s 0 100 50 2000 0.50 4'
thread4_exp4 = '0.50 2519.5 0.0688s 0 100 50 4000 0.50 4'
thread4_exp5 = '0.50 2523.9 0.1681s 0 100 50 8000 0.50 4'
```

```python
exps = [
    thread4_exp1.split(),
    thread4_exp2.split(),
    thread4_exp3.split(),
    thread4_exp4.split(),
    thread4_exp5.split()
]
```

```python
time = np.array([float(i[2][:-1]) for i in exps])
```

```python
plt.figure(figsize=(10, 10))
plt.plot(N, time, 'ro')
plt.plot(N, time)
plt.grid()
plt.show()
```

S(N)

```python
thread1_exp1 = '0.52 2328.1 0.0186s 0 100 50 500 0.50 1'
thread4_exp1 = '0.54 2591.6 0.0147s 0 100 50 500 0.50 4'
thread1_exp2 = '0.50 2510.2 0.0410s 0 100 50 1000 0.50 1'
thread4_exp2 = '0.49 2472.1 0.0227s 0 100 50 1000 0.50 4'
thread1_exp3 = '0.50 2562.6 0.0750s 0 100 50 2000 0.50 1'
thread4_exp3 = '0.49 2463.7 0.0371s 0 100 50 2000 0.50 4'
thread1_exp4 = '0.50 2509.8 0.1506s 0 100 50 4000 0.50 1'
thread4_exp4 = '0.50 2457.7 0.0770s 0 100 50 4000 0.50 4'
thread1_exp5 = '0.50 2506.8 0.3061s 0 100 50 8000 0.50 1'
thread4_exp5 = '0.50 2523.0 0.1585s 0 100 50 8000 0.50 4'
```

```python
thread1_exps = np.array([
    thread1_exp1.split(), 
    thread1_exp2.split(), 
    thread1_exp3.split(), 
    thread1_exp4.split(), 
    thread1_exp5.split()])
thread4_exps = np.array([
    thread4_exp1.split(), 
    thread4_exp2.split(), 
    thread4_exp3.split(), 
    thread4_exp4.split(), 
    thread4_exp5.split()])
```

```python
S = np.array([float(i[2][:-1]) / float(j[2][:-1]) 
     for i, j in zip(thread1_exps, thread4_exps)])
```

```python
plt.figure(figsize=(10, 10))
plt.plot(N, S, 'ro')
plt.plot(N, S)
plt.grid()
plt.show()
```

E(N)

```python
plt.figure(figsize=(10, 10))
plt.plot(N, E, 'ro')
plt.plot(N, E)
plt.grid()
plt.show()
```

T(P), S(P), E(P)

N = 20000

```python
thread1 = '0.50 2511.0 0.7950s 0 100 50 20000 0.50 1'
thread2 = '0.50 2518.3 0.4310s 0 100 50 20000 0.50 2'
thread4 = '0.50 2512.5 0.3835s 0 100 50 20000 0.50 4'
thread8 = '0.50 2523.2 0.3256s 0 100 50 20000 0.50 8'
thread16 = '0.50 2524.2 0.3120s 0 100 50 20000 0.50 16'
```

```python
exps = [
    thread1.split(),
    thread2.split(),
    thread4.split(),
    thread8.split(),
    thread16.split()
]
```

```python
P = np.array([2 ** i for i in range(5)])
```

T(P)

```python
T = np.array([float(i[2][:-1]) for i in exps])
```

```python
plt.figure(figsize=(10, 10))
plt.plot(P, T, 'ro')
plt.plot(P, T)
plt.grid()
plt.show()
```

S(P)

```python
S = T[0] / T
```

```python
plt.figure(figsize=(10, 10))
plt.plot(P, S, 'ro')
plt.plot(P, S)
plt.grid()
plt.show()
```

E(P)

```python
E = S / P
```

```python
plt.figure(figsize=(10, 10))
plt.plot(P, E, 'ro')
plt.plot(P, E)
plt.grid()
plt.show()
```
