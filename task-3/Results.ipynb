```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

N = 1000000, M = 2000

line format: time of builtin sort, time of parallel sort, P

```python
exp1 = [0.16696, 0.14462, 1]
exp2 = [0.24815, 0.09247, 2]
exp3 = [0.20059, 0.07209, 4]
exp4 = [0.23826, 0.05531, 8]
exp5 = [0.23792, 0.04863, 16]
```

```python
exps = [exp1, exp2, exp3, exp4, exp5]
```

```python
T = np.array([i[1] for i in exps])
T_builtin = np.array([i[0] for i in exps])
P = np.array([i[2] for i in exps])
```

T(P)

```python
plt.figure(figsize=(10, 10))
plt.plot(P, T, 'ro')
plt.plot(P, T)
plt.plot(P, T_builtin, 'bo')
plt.plot(P, T_builtin)
plt.grid()
plt.show()
```

S(P)

```python
S = T[0] / T
```

```python
plt.figure(figsize=(10, 10))
plt.plot(P, S, 'ro')
plt.plot(P, S)
plt.grid()
plt.show()
```

E(P)

```python
E = S / P
```

```python
plt.figure(figsize=(10, 10))
plt.plot(P, E, 'ro')
plt.plot(P, E)
plt.grid()
plt.show()
```
