```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

N = 20000000 M = 5000 

line format: time of builtin sort, time of parallel sort, P

```python
exp1 = [6.75371, 6.66982, 1]
exp2 = [6.28746, 4.11054, 2]
exp3 = [6.37079, 2.96823, 4]
exp4 = [6.20730, 3.21434, 8]
exp5 = [6.16118, 3.00052, 16]
```

```python
exps = [exp1, exp2, exp3, exp4, exp5]
```

```python
T = np.array([i[1] for i in exps])
T_builtin = np.array([i[0] for i in exps])
P = np.array([i[2] for i in exps])
```

T(P)

```python
plt.figure(figsize=(10, 10))
plt.plot(P, T, 'bo')
plt.plot(P, T, label="parallel sort time")
plt.plot(P, T_builtin, 'ro')
plt.plot(P, T_builtin, label="builtin sort time")
plt.xlabel("number of threads P")
plt.ylabel("work rime T")
plt.grid()
plt.legend()
plt.show()
```

S(P)

```python
S = T[0] / T
```

```python
plt.figure(figsize=(10, 10))
plt.plot(P, S, 'ro')
plt.plot(P, S)
plt.xlabel("number of threads P")
plt.ylabel("speedup S")
plt.grid()
plt.show()
```

E(P)

```python
E = S / P
```

```python
plt.figure(figsize=(10, 10))
plt.plot(P, E, 'ro')
plt.plot(P, E)
plt.xlabel("number of threads P")
plt.ylabel("efficiency E")
plt.grid()
plt.show()
```
